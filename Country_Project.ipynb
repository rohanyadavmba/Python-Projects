**Percentile of our data**

```python

twe=np.percentile(data["Population in thousands (2017)"] ,25)
fif=np.percentile(data["Population in thousands (2017)"],50)
sve=np.percentile(data["Population in thousands (2017)"],75)
```

```python
print(twe,fif,sve)
```

```python
# Quartiles
# Q1- 25
# Q2- 50
# Q3- 75
```

```python
q1=data["Population in thousands (2017)"].quantile(0.25)
print(q1)

q2=data["Population in thousands (2017)"].quantile(0.50)
print(q2)

q3=data["Population in thousands (2017)"].quantile(0.75)
print(q3)
```

```python
# IQR range -

iqr=q3-q1
print(iqr)
```

**Lowerfence** - The value below which all other values will become Outliers

```python
# lower_fence=q1-(k*iqr)  - Value of k is 1.5 or in some site you will find k=3 , k is a constant that determines how far away from the quartiles the fence should be
lf=q1-(1.5*iqr)
print(lf)
```

**Upperfence** - The value above which all other values will become Outliers

```python
uf=q3+(1.5*iqr)
print(uf)
# q3- 19193.0
# iqr - 18762.0
```

**Outliers** - Any value less than -27712.0 and above 47336.0 will be treated as Outliers

```python
# lets find which values are our outliers

data[ (data["Population in thousands (2017)"] >uf) | (data["Population in thousands (2017)"] <lf )]
```
